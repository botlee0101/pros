```python
from PIL import Image
import imagehash
import os
 
def calculate_hash(image_path):
    # Calculate the perceptual hash of the image
    img = Image.open(image_path)
    return imagehash.average_hash(img)
 
def delete_similar_images(folder_path, threshold=5):
    # Get a list of all image files in the folder
    image_files = [f for f in os.listdir(folder_path) if f.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp'))]
 
    # Dictionary to store hashes and corresponding file paths
    hash_dict = {}
 
    # Calculate and store hashes
    for image_file in image_files:
        image_path = os.path.join(folder_path, image_file)
        img_hash = calculate_hash(image_path)
        hash_dict.setdefault(img_hash, []).append(image_path)
 
    # Identify and delete similar images
    for hash_key, file_paths in hash_dict.items():
        if len(file_paths) > 1:
            print(f"Similar images found: {file_paths}")
            for file_path in file_paths[1:]:
                os.remove(file_path)
                print(f"Deleted: {file_path}")
 
if __name__ == "__main__":
    folder_path = r"C:\Users\SagarghotekarKalu-Co\Desktop\New folder\2D"
    delete_similar_images(folder_path)
```
